Exercício 1

```python

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

M = 1000

x = np.random.uniform(0, 1, (M, 1))
w = np.random.randn(M, 1)

y = 1.2 + 2.3*x + 10*w

plt.figure(figsize=(8,5))
plt.plot(x, y, '.b')
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.grid()
plt.savefig('exercicio_1_figura_2D.png')
plt.show()
```

Exercício 2

```python
x = np.random.uniform(0, 1, (M, 1))
y = np.random.uniform(0, 1, (M, 1))
z = x + y

plt.figure(figsize=(8,5))
plt.hist(z.ravel(), bins=30, density=True, edgecolor='black')
plt.xlabel('z', fontsize=14)
plt.ylabel('Densidade', fontsize=14)
plt.title('Exercício 2 - Histograma normalizado de z = x + y', fontsize=14)
plt.grid(alpha=0.3)


plt.savefig('exercicio_2_histograma_normalizado.png')

plt.show()
```

Exercício 3

```python
x1 = np.arange(-10, 11.25, 0.25)
x2 = np.arange(-10, 11.25, 0.25)
X1, X2 = np.meshgrid(x1, x2)
Y = X1**2 + X2**2

ax = plt.figure(figsize=(8,6)).add_subplot(projection='3d')
ax.plot_surface(X1, X2, Y, cmap='viridis')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
plt.show()
```
